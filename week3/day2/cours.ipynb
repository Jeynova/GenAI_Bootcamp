```python
import numpy as np

def generate_random_matrix(rows, cols, min_val=0, max_val=10, dtype=int, seed=None):
    """
    GÃ©nÃ¨re une matrice alÃ©atoire NumPy avec les dimensions et la plage spÃ©cifiÃ©es.

    Parameters:
    - rows (int): nombre de lignes
    - cols (int): nombre de colonnes
    - min_val (int|float): valeur minimale (incluse)
    - max_val (int|float): valeur maximale
    - dtype (type): int ou float
    - seed (int|None): graine alÃ©atoire (facultatif)

    Returns:
    - np.ndarray: matrice de forme (rows, cols)
    """
    if seed is not None:
        np.random.seed(seed)

    if dtype == int:
        return np.random.randint(min_val, max_val, size=(rows, cols))
    elif dtype == float:
        return np.random.uniform(min_val, max_val, size=(rows, cols))
    else:
        raise ValueError("dtype must be int or float")

# Exemple d'utilisation
matrix = generate_random_matrix(3, 5, min_val=0, max_val=10, dtype=int, seed=42)
print("Matrice alÃ©atoire :\n", matrix)
```

ðŸ” 1. Rolling Statistics â€” Statistiques glissantes
ðŸ§  DÃ©finition :

    Ce sont des statistiques calculÃ©es sur une fenÃªtre mobile dans une sÃ©rie temporelle (temps, date, index).

Tu dÃ©finis une fenÃªtre de taille fixe (ex : 3 jours), qui â€œglisseâ€ le long de ta sÃ©rie.

```python

# SÃ©rie temporelle fictive
s = pd.Series([10, 12, 15, 14, 18, 22, 25])

# Moyenne mobile sur 3 valeurs
rolling_mean = s.rolling(window=3).mean()

print(rolling_mean)
```

ðŸ” Le comportement par dÃ©faut de .rolling(window=N)

    Quand tu fais rolling(window=3).mean(), Pandas prend un bloc de 3 valeurs consÃ©cutives, et calcule la moyenne Ã  partir du moment oÃ¹ il a assez de valeurs.

âš ï¸ Par dÃ©faut, le calcul est :

    non centrÃ© (la fenÃªtre regarde en arriÃ¨re),

    basÃ© sur les 3 derniÃ¨res lignes, câ€™est-Ã -dire :
    mean(i)=moyenne des valeurs [iâˆ’2,iâˆ’1,i]
    mean(i)=moyenne des valeurs [iâˆ’2,iâˆ’1,i]

ðŸ” Exemple que tu as donnÃ© :

print(rolling_mean)

RÃ©sultat :
Index	Valeur	FenÃªtre utilisÃ©e	Moyenne
0	10	Pas assez de valeurs	NaN
1	12	Pas assez de valeurs	NaN
2	15	[10, 12, 15]	(10+12+15)/3 = 12.33
3	14	[12, 15, 14]	13.66
4	18	[15, 14, 18]	15.66
5	22	[14, 18, 22]	18.00
6	25	[18, 22, 25]	21.66
âœ… RÃ©ponse Ã  ta question :

    "On commence le mean au 3e ou au 4e jour ? Et on prend les 3 derniers ou tous Ã  partir du 3e ?"

    âœ… On commence Ã  calculer la moyenne au 3e Ã©lÃ©ment (index 2), quand il y a 3 valeurs disponibles.

    âœ… On prend les 3 derniÃ¨res valeurs, pas toutes depuis le dÃ©but.

    âŒ On ne prend pas toutes les valeurs Ã  partir du 3e jour. Ce sont des fenÃªtres glissantes (rolling = roulant/glissant).

ðŸªŸ 2. Window Functions â€” Fonctions de fenÃªtre (gÃ©nÃ©ralisÃ©es)
ðŸ§  DÃ©finition :

    Ce sont des fonctions qui appliquent des calculs sur un groupe ou une plage spÃ©cifique dans un ensemble de donnÃ©es.

Elles sont plus gÃ©nÃ©rales que les rolling statistics :

    Tu peux dÃ©finir des fenÃªtres centrÃ©es, dÃ©calÃ©es, pondÃ©rÃ©es, cumulatives,

ðŸ”„ DiffÃ©rence :

    rolling = fenÃªtre glissante, souvent sur des lignes consÃ©cutives (temps).

ðŸ’¡ Option : fenÃªtre centrÃ©e

ðŸ§® 2. Window functions (via .groupby().transform() ou .expanding(), .apply()...) = plus gÃ©nÃ©ral

```python
# Creating a large DataFrame
data = pd.DataFrame(np.random.rand(100000, 4), columns=['A', 'B', 'C', 'D'])

# Timing Pandas operations
start_time = time.time()
result_pandas = data[(data['A'] > 0.5) & (data['B'] < 0.5)].mean()
end_time = time.time()
print("Pandas Conditional Mean:\n", result_pandas)
print("Pandas Operation Time: {:.5f} seconds".format(end_time - start_time))

# Timing equivalent NumPy operations
numpy_data = data.to_numpy()
start_time = time.time()
result_numpy = np.mean(numpy_data[(numpy_data[:, 0] > 0.5) & (numpy_data[:, 1] < 0.5)], axis=0)
end_time = time.time()
print("NumPy Conditional Mean:\n", result_numpy)
print("NumPy Operation Time: {:.5f} seconds".format(end_time - start_time))
# Plotting with NumPy arrays
plt.scatter(data['A'], data['B'], alpha=0.5)
plt.title("Custom Scatter Plot")
plt.xlabel("A")
plt.ylabel("B")
plt.show()
```
