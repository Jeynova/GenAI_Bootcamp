```python
import numpy as np

def generate_random_matrix(rows, cols, min_val=0, max_val=10, dtype=int, seed=None):
    """
    Génère une matrice aléatoire NumPy avec les dimensions et la plage spécifiées.

    Parameters:
    - rows (int): nombre de lignes
    - cols (int): nombre de colonnes
    - min_val (int|float): valeur minimale (incluse)
    - max_val (int|float): valeur maximale
    - dtype (type): int ou float
    - seed (int|None): graine aléatoire (facultatif)

    Returns:
    - np.ndarray: matrice de forme (rows, cols)
    """
    if seed is not None:
        np.random.seed(seed)

    if dtype == int:
        return np.random.randint(min_val, max_val, size=(rows, cols))
    elif dtype == float:
        return np.random.uniform(min_val, max_val, size=(rows, cols))
    else:
        raise ValueError("dtype must be int or float")

# Exemple d'utilisation
matrix = generate_random_matrix(3, 5, min_val=0, max_val=10, dtype=int, seed=42)
print("Matrice aléatoire :\n", matrix)
```

🔍 1. Rolling Statistics — Statistiques glissantes
🧠 Définition :

    Ce sont des statistiques calculées sur une fenêtre mobile dans une série temporelle (temps, date, index).

Tu définis une fenêtre de taille fixe (ex : 3 jours), qui “glisse” le long de ta série.
À chaque position, tu calcules une statistique (ex : moyenne, écart-type…).
📈 Objectif :

    Lisser les données brutes (souvent bruyantes),

    Voir les tendances au lieu des pics brusques,

    Exemples : moyenne mobile, écart-type glissant, somme glissante.

```python

# Série temporelle fictive
s = pd.Series([10, 12, 15, 14, 18, 22, 25])

# Moyenne mobile sur 3 valeurs
rolling_mean = s.rolling(window=3).mean()

print(rolling_mean)
```

🔍 Le comportement par défaut de .rolling(window=N)

    Quand tu fais rolling(window=3).mean(), Pandas prend un bloc de 3 valeurs consécutives, et calcule la moyenne à partir du moment où il a assez de valeurs.

⚠️ Par défaut, le calcul est :

    non centré (la fenêtre regarde en arrière),

    basé sur les 3 dernières lignes, c’est-à-dire :
    mean(i)=moyenne des valeurs [i−2,i−1,i]
    mean(i)=moyenne des valeurs [i−2,i−1,i]

🔍 Exemple que tu as donné :

s = pd.Series([10, 12, 15, 14, 18, 22, 25])
rolling_mean = s.rolling(window=3).mean()
print(rolling_mean)

Résultat :
Index	Valeur	Fenêtre utilisée	Moyenne
0	10	Pas assez de valeurs	NaN
1	12	Pas assez de valeurs	NaN
2	15	[10, 12, 15]	(10+12+15)/3 = 12.33
3	14	[12, 15, 14]	13.66
4	18	[15, 14, 18]	15.66
5	22	[14, 18, 22]	18.00
6	25	[18, 22, 25]	21.66
✅ Réponse à ta question :

    "On commence le mean au 3e ou au 4e jour ? Et on prend les 3 derniers ou tous à partir du 3e ?"

    ✅ On commence à calculer la moyenne au 3e élément (index 2), quand il y a 3 valeurs disponibles.

    ✅ On prend les 3 dernières valeurs, pas toutes depuis le début.

    ❌ On ne prend pas toutes les valeurs à partir du 3e jour. Ce sont des fenêtres glissantes (rolling = roulant/glissant).

🪟 2. Window Functions — Fonctions de fenêtre (généralisées)
🧠 Définition :

    Ce sont des fonctions qui appliquent des calculs sur un groupe ou une plage spécifique dans un ensemble de données.

Elles sont plus générales que les rolling statistics :

    Tu peux définir des fenêtres centrées, décalées, pondérées, cumulatives,

    Tu peux appliquer des fonctions personnalisées (pas juste des stats standards).

🔄 Différence :

    rolling = fenêtre glissante, souvent sur des lignes consécutives (temps).

    window = peut être plus complexe (centrée, sur des groupes, sur des conditions).

💡 Option : fenêtre centrée

Si tu veux une moyenne centrée autour de l’élément courant, tu peux ajouter center=True :

s.rolling(window=3, center=True).mean()

Cela prendra [i-1, i, i+1] au lieu de [i-2, i-1, i].

📊 1. rolling() = fenêtre mobile régulière (index consécutifs)

    Utilisée surtout pour les séries temporelles.

    Fenêtre de taille fixe qui glisse ligne par ligne.

    Limité à des fonctions statistiques simples (mean, std, min, max, etc.).

    Peut être centrée ou non (center=True ou non).

    Index non-groupé, en général linéaire.

```python
# Médiane glissante sur 3 points
s.rolling(window=3).apply(lambda x: np.median(x))
```

🧮 2. Window functions (via .groupby().transform() ou .expanding(), .apply()...) = plus général

    Fenêtres par groupe, par conditions, ou expansives (toutes les valeurs jusqu’à un point donné).

    Supporte des fonctions plus complexes (lambda, fonctions personnalisées).

    Utilisable dans un contexte groupé (groupby(), expanding(), etc.).

    Peut simuler un rolling, mais avec plus de contrôle sur la forme et la logique.

➡️ Exemple :

df['group_mean'] = df.groupby('categorie')['valeurs'].transform(lambda x: x.rolling(3).mean())

✅ Résumé pédagogique
Terme	But principal	Exemple
Rolling Statistics	Lisser des séries temporelles	Moyenne mobile sur 7 jours
Window Functions	Appliquer des calculs sur des sous-ensembles de lignes	Médiane glissante, pondérée, ou personnalisée


🧠 À retenir pour toi :

    rolling est un cas particulier de window function.

    Utilise rolling quand tu veux faire des moyennes mobiles simples.

    Utilise les window functions (via rolling().apply() ou groupby().transform()) pour aller plus loin.

    📌 En résumé :

Critère	                        rolling()	                                    window functions (via transform, apply)
Usage principal	                Séries temporelles	Calculs groupés,            conditions complexes
Type de fenêtre	                Fixe,glissante	                                Fixe, glissante, ou personnalisée
Peut s’appliquer par groupe ?	❌ (sauf si imbriqué dans groupby)	          ✅ (via groupby().transform())
Fonctions supportées	        Simples (mean, std, sum…)	                    Très flexibles (lambda, UDF…)
Centrage possible ?	            Oui (center=True)	                            Oui si tu l’implémentes

🧪 Exemple comparatif :
Rolling :

df['moyenne_mobile'] = df['valeurs'].rolling(window=3).mean()

Window par groupe :

df['moyenne_par_groupe'] = df.groupby('categorie')['valeurs'].transform(lambda x: x.rolling(3).mean())

```python
dates = pd.date_range('2022-01-01', '2022-12-31', freq='D')
""" print("Date Range:\n", dates) """
temp_avg = np.random.uniform(15, 35, len(dates))
""" print("Average Temperatures:\n", temp_avg) """
humidity_avg = np.random.uniform(30, 90, len(dates))
""" print("Average Humidity:\n", humidity_avg)   """
df = pd.DataFrame({
    'Date': dates,
    'Temperature': temp_avg,
    'Humidity': humidity_avg
})
""" print("DataFrame:\n", df) """
temp_median = df['Temperature'].rolling(window=7).median()
""" print("7-Day Rolling Median of Temperature:\n", temp_median) """
hum_median = df['Humidity'].rolling(window=7).median()
""" print("7-Day Rolling Median of Humidity:\n", hum_median) """
temp_mean = df['Temperature'].rolling(window=7).mean()
""" print("7-Day Rolling Mean of Temperature:\n", temp_mean) """
hum_mean = df['Humidity'].rolling(window=7).mean()
""" print("7-Day Rolling Mean of Humidity:\n", hum_mean) """
temp_std = df['Temperature'].rolling(window=7).std()
""" print("7-Day Rolling Standard Deviation of Temperature:\n", temp_std) """
hum_std = df['Humidity'].rolling(window=7).std()
""" print("7-Day Rolling Standard Deviation of Humidity:\n", hum_std) """
highest_temp = df['Temperature'].rolling(window=7).max()
""" print("7-Day Rolling Maximum Temperature:\n", highest_temp) """
highest_hum = df['Humidity'].rolling(window=7).max()
""" print("7-Day Rolling Maximum Humidity:\n", highest_hum) """
# Example of a rolling window function
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Temperature'], label='Temperature', color='orange')
plt.plot(df['Date'], temp_mean, label='7-Day Rolling Mean', color='blue')
plt.fill_between(df['Date'], temp_mean - temp_std, temp_mean + temp_std, color='lightblue', alpha=0.5, label='±1 Std Dev')
plt.title('Temperature with Rolling Mean and Std Dev')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Creating a large DataFrame
data = pd.DataFrame(np.random.rand(100000, 4), columns=['A', 'B', 'C', 'D'])

# Timing Pandas operations
start_time = time.time()
result_pandas = data[(data['A'] > 0.5) & (data['B'] < 0.5)].mean()
end_time = time.time()
print("Pandas Conditional Mean:\n", result_pandas)
print("Pandas Operation Time: {:.5f} seconds".format(end_time - start_time))

# Timing equivalent NumPy operations
numpy_data = data.to_numpy()
start_time = time.time()
result_numpy = np.mean(numpy_data[(numpy_data[:, 0] > 0.5) & (numpy_data[:, 1] < 0.5)], axis=0)
end_time = time.time()
print("NumPy Conditional Mean:\n", result_numpy)
print("NumPy Operation Time: {:.5f} seconds".format(end_time - start_time))
# Plotting with NumPy arrays
plt.scatter(data['A'], data['B'], alpha=0.5)
plt.title("Custom Scatter Plot")
plt.xlabel("A")
plt.ylabel("B")
plt.show()
```
