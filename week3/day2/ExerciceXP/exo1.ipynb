```python
# generer 30 scores avant formation autour de 50
# utiliser une graine pour garder les memes valeurs
np.random.seed(0)
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# simuler scores apres formation en ajoutant en moyenne 5 points
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# h0 la formation ne change rien
# h1 la formation augmente la productivite

# moyenne avant
mean_before = np.mean(productivity_before)
# moyenne apres
mean_after = np.mean(productivity_after)

# ecart type avant
std_before = np.std(productivity_before)
# ecart type apres
std_after = np.std(productivity_after)

# difference entre les deux moyennes
diff = mean_after - mean_before

# afficher resultats
print(f"moyenne avant : {mean_before:.2f}")
print(f"moyenne apres : {mean_after:.2f}")
print(f"difference moyenne : {diff:.2f}")

# test t pour comparer les memes personnes avant et apres
stat, p_value = ttest_rel(productivity_after, productivity_before)
print(f"statistique t = {stat:.2f}, p-value = {p_value:.4f}")
```

```python

# creer une serie de dates pour toute l annee 2023
dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")

# generer des valeurs entre -5 et 35 pour chaque jour
values = np.random.uniform(low=-5, high=35, size=len(dates))

# creer un dataframe avec les dates comme index
df = pd.DataFrame({'Date': dates, 'Value': values}).set_index('Date')

# definir les trimestres avec leur date de debut et de fin
quarters = {
    "Janvier - Mars": ("2023-01-01", "2023-03-31"),
    "Avril - Juin": ("2023-04-01", "2023-06-30"),
    "Juillet - Septembre": ("2023-07-01", "2023-09-30"),
    "Octobre - Decembre": ("2023-10-01", "2023-12-31")
}

# afficher les premieres lignes de chaque trimestre
for name, (start, end) in quarters.items():
    data = df[start:end]
    print(f"\n{name} :")
    print(data.head())

# preparer une figure avec 4 graphiques
plt.figure(figsize=(12, 8))
colors = ['blue', 'green', 'orange', 'red']

# afficher chaque trimestre dans un graphique
for i, (name, (start, end)) in enumerate(quarters.items(), 1):
    data = df[start:end]
    plt.subplot(2, 2, i)
    plt.plot(data.index, data['Value'], color=colors[i-1])
    plt.title(name)
    plt.xlabel("Date")
    plt.xticks(rotation=45)
    plt.ylabel("Valeur")
    plt.grid(True)

# ajuster les espaces entre les graphiques
plt.tight_layout()
# afficher la figure
plt.show()
```
