```python
df = np.random.rand(100, 5)
df = pd.DataFrame(df, columns=['A', 'B', 'C', 'D', 'E'])
df['Group'] = np.random.choice(['X', 'Y'], size=100)
# Display the first few rows of the DataFrame
print(df.head())

plt.figure(figsize=(10, 6))
plt.hist(df['A'], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Column A')
plt.xlabel('Values of A')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.savefig('histogram_A.png')
plt.show()
```


üîπ Niveau 2 : Statistiques descriptives

    Cr√©e un DataFrame Pandas avec deux colonnes : group (cat√©gorie A ou B), value (valeurs num√©riques)

    Calcule moyenne, m√©diane, std par groupe

    Affiche un boxplot par groupe

```python
df = pd.DataFrame(columns=['Group', 'Value'])  # ‚úîÔ∏è Deux colonnes vides

df['Group'] = np.random.choice(['A', 'B'], size=100)
df['Value'] = np.random.rand(100)

# creer le dataframe directement avec un dictionnaire
df2 = pd.DataFrame({
    'Group': np.random.choice(['A', 'B'], size=100),
    'Value': np.random.rand(100)
})
df_means = df.groupby('Group')['Value'].mean().reset_index()
print(df_means)
df_middle = df.groupby('Group')['Value'].median().reset_index()
print(df_middle)
df_std = df.groupby('Group')['Value'].std().reset_index()
print(df_std)
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(x='Group', y='Value', data=df, palette='Set2')
boxplot.set_title('Boxplot of Values by Group')
plt.savefig('boxplot_group_values.png')
plt.show()
```
