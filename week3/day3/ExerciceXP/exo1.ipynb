```python
import numpy as np
from scipy import stats
import scipy as sp
import matplotlib.pyplot as plt
from scipy.stats import binom,norm,poisson,expon,pearsonr, spearmanr
import seaborn as sns
import pandas as pd
```

```python
print("SciPy version:", sp.__version__)
```

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_scipy = stats.tmean(data)
median_scipy = np.median(data) 
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)

print("Mean:", mean_scipy)
print("Median:", median_scipy)
print("Variance:", variance_scipy)
print("Standard Deviation:", std_dev_scipy )
```

```python
mean = 50
std_dev = 10

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

# Tracer la distribution
plt.plot(x, y)
plt.title("Distribution Normale (mean = 50, std_dev = 10)")
plt.xlabel("Valeur")
plt.ylabel("Densit√© de probabilit√©")
plt.grid(True)
plt.show()
```

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_val = stats.ttest_ind(data1, data2)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Il y a une diff√©rence significative entre les deux groupes.")
else:
    print("Aucune diff√©rence significative d√©tect√©e.")
```

```python
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
predicted_price = slope * 90 + intercept

print("Pente (slope):", slope)
print("Ordonn√©e √† l'origine (intercept):", intercept)
print(f"Prix pr√©dit pour une maison de 90 m¬≤: {predicted_price:.2f}")


plt.scatter(house_sizes, house_prices, color='blue', label='Donn√©es')
plt.plot(house_sizes, slope * house_sizes + intercept, color='red', label='R√©gression lin√©aire')
plt.xlabel('Taille de la maison (m¬≤)')
plt.ylabel('Prix de la maison')
plt.title("R√©gression lin√©aire entre taille de la maison et prix")
plt.legend()
plt.grid(True)
plt.show()

print(f"La pente indique qu'√† chaque augmentation de 1 m¬≤, le prix de la maison augmente de {slope:.2f} ‚Ç¨.")
```

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_stat, p_val = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-statistic:", f_stat)
print("P-value:", p_val)

# Interpr√©tation du P-value
if p_val < 0.05:
    print("Les engrais ont un effet significatif sur la croissance des plantes.")
else:
    print("Il n'y a pas d'effet significatif des engrais sur la croissance des plantes.")

# Visualisation des donn√©es avec un boxplot
data = [fertilizer_1, fertilizer_2, fertilizer_3]
labels = ['Fertilizer 1', 'Fertilizer 2', 'Fertilizer 3']

plt.figure(figsize=(8, 6))
plt.boxplot(data, tick_labels=labels)
plt.title("Comparaison de la croissance des plantes en fonction de l'engrais")
plt.xlabel("Type d'engrais")
plt.ylabel("Croissance des plantes (cm)")
plt.grid(True)
plt.show()
```

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

pearson_corr, _ = pearsonr(data['age'], data['income'])
print("Coefficient de corr√©lation de Pearson:", pearson_corr)

spearman_corr, _ = spearmanr(data['age'], data['income'])
print("Coefficient de corr√©lation de Spearman:", spearman_corr)
```
