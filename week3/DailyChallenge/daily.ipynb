```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities_df = pd.read_csv("cities.csv")
selected_cities = cities_df["city_name"].dropna().unique()[:10]

temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12)) 
print(temperature_data)
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_temps = pd.DataFrame(temperature_data, index=selected_cities, columns=months)
print(df_temps)
```

```python
annual_avg = df_temps.mean(axis=1)
max_city = annual_avg.idxmax()
min_city = annual_avg.idxmin()

print("Moyenne annuelle par ville :")
print(annual_avg)
print(f"\nVille la plus chaude : {max_city} ({annual_avg[max_city]:.2f}¬∞C)")
print(f"Ville la plus froide : {min_city} ({annual_avg[min_city]:.2f}¬∞C)")
```

```python
print(annual_avg)
annual_avg.max() - annual_avg.min()
```

```python
plt.figure(figsize=(12, 6))
for city in df_temps.index:
    plt.plot(months, df_temps.loc[city], label=city)

plt.title("Temp√©ratures mensuelles moyennes par ville")
plt.xlabel("Mois")
plt.ylabel("Temp√©rature (¬∞C)")
plt.legend(loc="upper right")
plt.grid(True)
plt.tight_layout()
plt.show()
```
