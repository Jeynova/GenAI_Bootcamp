```python
import random
import time
from IPython.display import clear_output
```

```python
def welcome():
    clear_output(wait=True)
    print(r"""
  _______  ___________________  _______   _  ____  ____  ______  _______  ____
 / ___/ / / / __/ __/ __/  _/ |/ / ___/  / |/ / / / /  |/  / _ )/ __/ _ \/ __/
/ (_ / /_/ / _/_\ \_\ \_/ //    / (_ /  /    / /_/ / /|_/ / _  / _// , _/\ \  
\___/\____/___/___/___/___/_/|_/\___/  /_/|_/\____/_/  /_/____/___/_/|_/___/  
                                                                                                                                                                   
    """)
    print("Cr√©√© pour le Crash Course Python")
    print("¬© 2025 - J√©r√©my Novice\n")
    input("Appuyez sur Entr√©e pour commencer...")
    clear_output(wait=True)

def difficulty():
    while True:
        choice = input("Choisissez une difficult√© (facile, moyen, difficile) ou 'q' pour quitter : ").strip().lower()
        if choice in ("facile", "f"):
            return 1, 50, 10
        elif choice in ("moyen", "m"):
            return 1, 100, 7
        elif choice in ("difficile", "d"):
            return 1, 200, 5
        elif choice in ("quitter", "q"):
            return None
        else:
            print("Choix invalide. Veuillez entrer 'facile', 'moyen' ou 'difficile'.")

def inputNum(attempt, min_val, max_val):
    try:
        return int(input(f"Essai {attempt} - Entrez un nombre entre {min_val} et {max_val} : "))
    except ValueError:
        print("Veuillez entrer un nombre valide.")
        return None

def clues(guess, target, min_val, max_val):
    if guess == 42:
        print(".-.. . / --. ..- .. -.. . / -.. ..- / ...- --- -.-- .- --. . ..- .-. / --. .- .-.. .- -.-. - .. --.- ..- .")
    if guess < target:
        print("Trop bas !")
        return max(min_val, guess + 1), max_val
    elif guess > target:
        print("Trop haut !")
        return min_val, min(max_val, guess - 1)
    else:
        return min_val, max_val

def calculate_score(attempt, max_attempts, duration):
    base_score = 1000
    difficulty_penalty = {10: 0, 7: 100, 5: 200}[max_attempts]
    score = base_score - (attempt - 1) * 50 - difficulty_penalty
    return max(score, 0), round(duration, 2)

def display_result(guess, target, attempt, max_attempts, duration):
    if guess == target:
        score, time_taken = calculate_score(attempt, max_attempts, duration)
        print(f"\nBravo ! Vous avez trouv√© {target} en {attempt} essai{'s' if attempt > 1 else ''}.")
        print(f"Temps √©coul√© : {time_taken} secondes")
        print(f"Score final : {score} points\n")
    else:
        print(f"Vous avez utilis√© tous vos essais. Le nombre √©tait {target}.\n")

def check_duplicate(guess, history):
    if guess in history:
        print("Vous avez d√©j√† essay√© ce nombre. Essayez un autre.")
        return True
    history.add(guess)
    return False

def give_hint(target):
    hints = []
    if target % 2 == 0:
        hints.append("Le nombre est pair.")
    else:
        hints.append("Le nombre est impair.")
    if target % 3 == 0:
        hints.append("Il est divisible par 3.")
    if target % 5 == 0:
        hints.append("Il est un multiple de 5.")
    if target % 10 == 0:
        hints.append("C'est un multiple de 10.")
    if target > 100:
        hints.append("Il est sup√©rieur √† 100.")
    elif target < 50:
        hints.append("Il est inf√©rieur √† 50.")
    return random.choice(hints) if hints else "Aucun indice disponible."

def final_hint(target, min_val, max_val, difficulty_level):
    import random

    options_count = {10: 3,7: 5,5: 7 }[difficulty_level]

    options = {target}
    while len(options) < options_count:
        n = random.randint(min_val, max_val)
        options.add(n)

    options = list(options)
    random.shuffle(options)
    return options

def play():
    difficulty_check = difficulty()
    if difficulty_check is None:
        print("Au revoir !")
        return False
    
    min_val, max_val, max_attempts = difficulty_check
    target = random.randint(min_val, max_val)
    attempts_history = set()
    start_time = time.time()

    print(f"\nVous avez {max_attempts} essais pour deviner un nombre entre {min_val} et {max_val}.")

    for attempt in range(1, max_attempts + 1):
        clear_output(wait=True)
        guess = inputNum(attempt, min_val, max_val)
        if attempt == max_attempts-1:
            options = final_hint(target, min_val, max_val, max_attempts)
            print("\nüí° Derni√®re chance ! Le nombre est parmi :", ', '.join(map(str, options)))

        if guess is None:
            continue

        if check_duplicate(guess, attempts_history):
            continue

        if guess == target:
            duration = time.time() - start_time
            display_result(guess, target, attempt, max_attempts, duration)
            break
        else:
            min_val, max_val = clues(guess, target, min_val, max_val)
            if attempt == max_attempts - 1:
                print("Indice :", give_hint(target))
            print("Essais pr√©c√©dents :", ', '.join(map(str, sorted(attempts_history))))
            
    else:
        duration = time.time() - start_time
        display_result(None, target, attempt, max_attempts, duration)

def loop():
    welcome()
    while True:
        if play() is False:
            break
        replay = input("Voulez-vous rejouer ? (o/n) : ").strip().lower()
        if replay != 'o':
            print("Merci d\'avoir jou√© ! √Ä bient√¥t")
            break

loop()
        

```
