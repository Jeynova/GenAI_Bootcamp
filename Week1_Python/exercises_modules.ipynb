```python
#Exercice 1
from typing import List

class Cat:
    is_lazy: bool = True

    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def walk(self):
        return f"{self.name} is just walking around"

    def __str__(self):
        return f"{self.name} is {self.age} years old"

class Bengal(Cat):
    def sing(self, sound: str):
        return f"{self.name} sings: {sound}"

class Chartreux(Cat):
    def sing(self, sound: str):
        return f"{self.name} sings: {sound}"

class Siamese(Cat):
    def __init__(self, name: str, age: int, eye_color: str = "blue"):
        super().__init__(name, age)
        self.eye_color = eye_color

    def sing(self, sound: str):
        return f"{self.name} sings beautifully: {sound}"

class Pets:
    def __init__(self, animals: List[Cat]):
        self.animals = animals

    def walk(self):
        for animal in self.animals:
            print(animal.walk())

cats: List[Cat] = [
    Bengal("Mistigri", 4),
    Chartreux("Felix", 6),
    Siamese("Chaussette", 3)
]

sara_pets = Pets(cats)
print("Cats going for a walk:")
sara_pets.walk()
```

```python
#Exercice 2

class Dog:
    def __init__(self, name: str, age: int, weight: int):
        if age <= 0 or weight <= 0:
            raise ValueError("Age and weight must be positive.")
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        return f"{self.name} is barking"

    def run_speed(self):
        return (self.weight / self.age) * 10

    def fight(self, other: "Dog"):
        my_power = self.run_speed() * self.weight
        their_power = other.run_speed() * other.weight
        if my_power > their_power:
            return f"{self.name} wins the fight"
        elif my_power < their_power:
            return f"{other.name} wins the fight"
        return "It's a draw"

dog1 = Dog("MEdor", 3, 20)
dog2 = Dog("Rex", 5, 25)
dog3 = Dog("Cubitus", 2, 15)

print(dog1.bark())
print(f"{dog2.name}'s run speed: {dog2.run_speed():.2f}")
print(dog1.fight(dog2))
```

```python
#Exercice 3
import random

class PetDog(Dog):
    def __init__(self, name: str, age: int, weight: int):
        super().__init__(name, age, weight)
        self.trained: bool = False

    def train(self):
        print(self.bark())
        self.trained = True

    def play(self, *others: "PetDog"):
        all_names = [self.name] + [dog.name for dog in others]
        print(f"{', '.join(all_names)} all play together!")

    def do_a_trick(self):
        if self.trained:
            tricks = [
                "does a barrel roll",
                "stands on his back legs",
                "shakes your hand",
                "plays dead"
            ]
            print(f"{self.name} {random.choice(tricks)}")
        else:
            print(f"{self.name} is not trained yet.")

pet_dog = PetDog("Rex", 4, 18)
pet_dog.train()
pet_dog.play(dog1, dog2)
pet_dog.do_a_trick()
```

```python
#Exercice 4
from dataclasses import dataclass
from typing import List

@dataclass
class Person:
    first_name: str
    age: int
    last_name: str = ""

    def is_18(self):
        return self.age >= 18

    def __str__(self):
        return f"{self.first_name} {self.last_name}, {self.age} ans"

class Family:
    def __init__(self, last_name: str):
        self.last_name = last_name
        self.members: List[Person] = []

    def born(self, first_name: str, age: int):
        if age < 0:
            raise ValueError("Age must be non-negative.")
        person = Person(first_name, age, self.last_name)
        self.members.append(person)

    def check_majority(self, first_name: str):
        person = next((m for m in self.members if m.first_name == first_name), None)
        if not person:
            print("Person not found.")
            return
        if person.is_18():
            print("You are over 18, your parents Jane and John accept that you will go out with your friends.")
        else:
            print("Sorry, you are not allowed to go out with your friends.")

    def family_presentation(self):
        print(f"The {self.last_name} family:")
        for member in self.members:
            print(f"- {member}")

# --- Test
fam = Family("Lambda")
fam.born("Alice", 17)
fam.born("Jeremy", 20)
fam.check_majority("Alice")
fam.check_majority("Jeremy")
fam.family_presentation()
```
