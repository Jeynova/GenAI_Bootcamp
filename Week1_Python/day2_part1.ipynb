üåü Exercise 8: Pizza Toppings

Key Python Topics:

    Loops
    Lists
    String formatting


Instructions:

    Write a loop that asks the user to enter pizza toppings one by one.
    Stop the loop when the user types 'quit'.
    For each topping entered, print:
    "Adding [topping] to your pizza."
    After exiting the loop, print all the toppings and the total cost of the pizza.
        The base price is $10, and each topping adds $2.50.


```python
# Exercice 1 : Listes en dictionnaire

def convert_lists_to_dict(keys_list, values_list):
    # Associe chaque cl√© √† une valeur de la m√™me position gr√¢ce √† zip
    result = dict(zip(keys_list, values_list))
    print("R√©sultat du dictionnaire :", result)
    return result

keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
convert_lists_to_dict(keys, values)


# Exercice 2 : Cin√©ma - calcul des tickets

def calculate_family_ticket_prices(family_ages):
    # Calcule le prix du billet selon l'√¢ge de chaque membre
    total_price = 0
    for name, age in family_ages.items():
        if age < 3:
            price = 0
        elif age <= 12:
            price = 10
        else:
            price = 15
        print(f"{name.capitalize()} : {price} ‚Ç¨")
        total_price += price
    print("Prix total :", total_price, "‚Ç¨")
    return total_price

family_ages = {"rick": 43, 'beth': 13, 'morty': 5, 'summer': 8}
calculate_family_ticket_prices(family_ages)

def input_family_and_calculate():
    # Demande les noms et √¢ges des membres jusqu'√† ce que l'utilisateur tape 'quitter
    family_ages = {}
    while True:
        name = input("Nom de la personne ou quitter: ")
        if name.lower() == 'quitter':
            break
        try:
            age = int(input(f"√Çge de {name} : "))
        except ValueError:
            print("Merci d'entrer un nombre pour l'√¢ge.")
            continue
        family_ages[name] = age
    if family_ages:
        calculate_family_ticket_prices(family_ages)
    else:
        print("Aucune donn√©e saisie.")

input_family_and_calculate()


# Exercice 3 : Dictionnaire Zara

def manipulate_zara_brand():
    # Cr√©ation et manipulation du dictionnaire avec les infos de la marque
    brand = {
        "name": "Zara",
        "creation_date": 1975,
        "creator_name": "Amancio Ortega Gaona",
        "type_of_clothes": ["men", "women", "children", "home"],
        "international_competitors": ["Gap", "H&M", "Benetton"],
        "number_stores": 7000,
        "major_color": {
            "France": "blue",
            "Spain": "red",
            "US": ["pink", "green"]
        }
    }
    brand["number_stores"] = 2
    print("Zara cible les clients suivants :", ", ".join(brand["type_of_clothes"]))
    brand["country_creation"] = "Spain"
    # Ajoute "Desigual" seulement si la cl√© existe
    if "international_competitors" in brand:
        brand["international_competitors"].append("Desigual")
    brand.pop("creation_date", None)
    print("Dernier concurrent international :", brand["international_competitors"][-1])
    print("Couleurs majeures aux US :", ", ".join(brand["major_color"]["US"]))
    print("Nombre de cl√©s dans le dictionnaire :", len(brand))
    print("Cl√©s du dictionnaire :", list(brand.keys()))
    # Bonus : fusion d'un autre dictionnaire
    more_on_zara = {"creation_date": 1975, "number_stores": 10000}
    brand.update(more_on_zara)
    print("Apr√®s fusion :", brand)

manipulate_zara_brand()


# Exercice 4 : D√©crire une ville

def describe_city(city, country="Inconnu"):
    # Affiche la ville et son pays (ou 'Inconnu' par d√©faut)
    print(f"{city} est en {country}.")

describe_city("Reykjavik", "Islande")
describe_city("Paris")


# Exercice 5 : Nombre al√©atoire et comparaison

import random

def compare_to_random(user_number):
    # G√©n√®re un nombre au hasard et compare avec la saisie utilisateur
    random_number = random.randint(1, 100)
    if user_number == random_number:
        print("Succ√®s !")
    else:
        print(f"√âchec ! Ton nombre : {user_number}, Nombre al√©atoire : {random_number}")

compare_to_random(42)


# Exercice 6 : T-shirt personnalis√©

def make_shirt(size="large", text="J'aime Python"):
    # Affiche les infos du t-shirt √† produire
    print(f"La taille du t-shirt est {size} et le texte est '{text}'.")

make_shirt()
make_shirt(size="medium")
make_shirt(size="small", text="Message personnalis√©")
make_shirt(text="Hello!", size="extra-large")


# Exercice 7 : Conseil m√©t√©o

def get_random_temp():
    # G√©n√®re une temp√©rature flottante entre -10 et 40
    return round(random.uniform(-10, 40), 1)

def main_temperature_advice():
    # Donne un conseil vestimentaire selon la temp√©rature
    temp = get_random_temp()
    print(f"The temperature right now is {temp}¬∞C.")
    if temp < 0:
        print("Brrr, that's freezing! Wear some extra layers today.")
    elif temp < 16:
        print("Quite chilly! Don't forget your coat.")
    elif temp < 24:
        print("Nice weather.")
    elif temp < 33:
        print("A bit warm, stay hydrated.")
    else:
        print("It's really hot! Stay cool.")

main_temperature_advice()


# Exercice 8 : Ingr√©dients pizza

def pizza_toppings():
    # Ajoute des ingr√©dients et calcule le prix total de la pizza
    toppings = []
    base_price = 10
    topping_price = 2.5
    while True:
        topping = input("Ajoute un ingr√©dient (ou tape 'quit' pour quitter) : ")
        if topping.lower() == 'quit':
            break
        print(f"Ajout de {topping} sur votre pizza.")
        toppings.append(topping)
    print("Ingr√©dients choisis :", ", ".join(toppings))
    total = base_price + topping_price * len(toppings)
    print(f"Prix total : {total} ‚Ç¨")

pizza_toppings()
```
