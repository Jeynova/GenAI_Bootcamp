```python
#Exercise 1: Cats

class Cat:
    def __init__(self, cat_name, cat_age):
        self.name = cat_name
        self.age = cat_age

# Instanciations
cat1 = Cat("Mistigri",12)
cat2 = Cat("Bolero",8)
cat3 = Cat("Felix",16)

# Fonction chat plus vieux
def find_oldest_cat(*cats):
    return max(cats, key=lambda cat: cat.age)

# Recherche et affichage du chat le plus âgé
oldest = find_oldest_cat(cat1, cat2, cat3)

# Essais
print(f"Le chat le plus âgé est {oldest.name}, âgé de {oldest.age} ans.")
```

```python
#Exercise 2
class Dog:
    def __init__(self, name: str, height: int):
        self.name = name
        self.height = height

    #Aboiement
    def bark(self) -> None:
        print(f"{self.name} aboie : WAF!!")
        
    #Afficher la hauter de saut
    def jump(self) -> None:
        print(f"{self.name} saute jusqu'à {self.height * 2} cm")

# Instanciations
dog1 = Dog("Rex", 50)
dog2 = Dog("Medor", 20)

#Tets methodes
dog1.bark()
dog1.jump()

dog2.bark()
dog2.jump()

# Comparaison des tailles
bigger_dog = dog1 if dog1.height > dog2.height else dog2
print(f"Le plus gros chien est : {bigger_dog.name}.")
```

```python
#Exercise 4

class Zoo:
    def __init__(self, zoo_name):
        self.name = zoo_name
        self.animals = list()
        self.groups = {}

    def add_animal(self, new_animal):
        if new_animal not in self.animals:
            self.animals.append(new_animal)
        # Remise à zero des groupe si ajout
        self.groups.clear()

    def get_animals(self):
        print('\n'.join(self.animals))

    def sell_animal(self, animal_sold):
        if animal_sold in self.animals:
            self.animals.remove(animal_sold)
        # Remise à zero des groupe si vente
        self.groups.clear()

    def sort_animals(self):
        #Trie et groupe les animaux
        if not self.animals:
            print("Aucun animal dans le zoo.")
            self.groups.clear()
            return {}

        sorted_animals = sorted(self.animals)
        groups = {}
        for animal in sorted_animals:
            letter = animal[0].upper()
            groups.setdefault(letter, []).append(animal)
        self.groups = groups
        return groups

    def get_groups(self):
        #Affiche les groupes
        if not self.groups:
            self.sort_animals()
        for k in sorted(self.groups.keys()):
            print(f"{k}: {self.groups[k]}")
            
ramat_gan_safari = Zoo("Ramat Gan Safari")

# Essais
ramat_gan_safari.add_animal("Giraffe")
ramat_gan_safari.add_animal("Bear")
ramat_gan_safari.add_animal("Baboon")
ramat_gan_safari.get_animals()
ramat_gan_safari.sell_animal("Bear")
ramat_gan_safari.get_animals()
ramat_gan_safari.sort_animals()
ramat_gan_safari.get_groups()
   
  
```
