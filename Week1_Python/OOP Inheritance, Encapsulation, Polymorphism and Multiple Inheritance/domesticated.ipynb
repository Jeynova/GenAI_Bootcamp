```python
#Exercice 3
import random
import dog

class PetDog(dog.Dog):
    def __init__(self, name: str, age: int, weight: int):
        super().__init__(name, age, weight)
        self.trained: bool = False

    def train(self):
        print(self.bark())
        self.trained = True

    def play(self, *others: "PetDog"):
        for other in others:
            if not isinstance(other, PetDog):
                raise TypeError("play() only accepts PetDog instances")
        all_names = [self.name] + [dog.name for dog in others]
        print(f"{', '.join(all_names)} all play together!")

    def do_a_trick(self):
        if self.trained:
            tricks = [
                "does a barrel roll",
                "stands on his back legs",
                "shakes your hand",
                "plays dead"
            ]
            print(f"{self.name} {random.choice(tricks)}")
        else:
            print(f"{self.name} is not trained yet.")
            


pet_dog1 = PetDog("Rex", 4, 18)
pet_dog2 = PetDog("Medor", 3, 16)
pet_dog1.train()
pet_dog1.play(pet_dog2)
pet_dog1.do_a_trick()
```
