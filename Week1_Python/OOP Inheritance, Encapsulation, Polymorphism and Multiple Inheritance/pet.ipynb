```python
#Exercice 1
from typing import List

class Cat:
    is_lazy: bool = True

    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def walk(self):
        return f"{self.name} is just walking around"

    def __str__(self):
        return f"{self.name} is {self.age} years old"

class Bengal(Cat):
    def sing(self, sound: str):
        return f"{self.name} sings: {sound}"

class Chartreux(Cat):
    def sing(self, sound: str):
        return f"{self.name} sings: {sound}"

class Siamese(Cat):
    def __init__(self, name: str, age: int, eye_color: str = "blue"):
        super().__init__(name, age)
        self.eye_color = eye_color

    def sing(self, sound: str):
        return f"{self.name} sings beautifully: {sound}"

class Pets:
    def __init__(self, animals: List[Cat]):
        self.animals = animals

    def walk(self):
        for animal in self.animals:
            print(animal.walk())

cats: List[Cat] = [
    Bengal("Mistigri", 4),
    Chartreux("Felix", 6),
    Siamese("Chaussette", 3)
]

sara_pets = Pets(cats)
print("Cats going for a walk:")
sara_pets.walk()
```
