```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Exercice 1
data = pd.read_csv("insurance.csv")

print("\n--- Apercu initial ---")
print(data.head())
print("\n--- Infos générales ---")
print(data.info())
print("\n--- Statistiques descriptives ---")
print(data.describe())
print("\n--- Valeurs manquantes ---")
print(data.isnull().sum())

# analyse
plt.figure(figsize=(8, 5))
sns.histplot(data['charges'], bins=30, kde=True)
plt.title("Distribution des charges")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x=data['smoker'], y=data['charges'])
plt.title("Charges selon le statut fumeur")
plt.show()

sns.scatterplot(x='age', y='charges', data=data, hue='smoker')
plt.title("Age vs Charges")
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Matrice de corrélation")
plt.show()

# Exercice 2
titanic = pd.read_csv("titanic.csv")

print("\n--- Apercu Titanic ---")
print(titanic.head())
print("\n--- Infos Titanic ---")
print(titanic.info())
print("\n--- Statistiques descriptives Titanic ---")
print(titanic.describe())
print("\n--- Valeurs manquantes Titanic ---")
print(titanic.isnull().sum())

# nettoyage
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic.drop(columns=['Cabin'], inplace=True)  # trop de valeurs manquantes

# transformation
titanic['Sex'] = titanic['Sex'].map({'female': 0, 'male': 1})

# analyse
sns.histplot(titanic['Age'], bins=30, kde=True)
plt.title("Distribution des âges")
plt.show()

sns.countplot(x='Survived', data=titanic)
plt.title("Répartition des survivants")
plt.show()

sns.boxplot(x='Pclass', y='Age', data=titanic)
plt.title("Classe vs Age")
plt.show()

sns.barplot(x='Sex', y='Survived', data=titanic)
plt.title("Sexe vs Survie")
plt.show()

sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Corrélation Titanic")
plt.show()

# Exercice 3
iris = pd.read_csv("Iris.csv")
print("\n--- Apercu Iris ---")
print(iris.head())
print("\n--- Infos Iris ---")
print(iris.info())
print("\n--- Valeurs manquantes Iris ---")
print(iris.isnull().sum())

sns.histplot(data=iris, x='SepalLengthCm', hue='Species', kde=True)
plt.title("Distribution SepalLength")
plt.show()

sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=iris)
plt.title("Petal Length vs Width")
plt.show()

sns.pairplot(iris, hue='Species')
plt.show()

sns.heatmap(iris.drop(columns=['Id']).corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Iris")
plt.show()
```
