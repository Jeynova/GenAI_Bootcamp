```python
import pandas as pd
```

```python
# Exercice 1

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

```python
# Exercice 2

divisible_by_3 = [x for x in range(1, 21) if x % 3 == 0]
print(divisible_by_3)
```

```python
# Exercice 4

boston = pd.read_csv("BostonHousing.csv")

mean_rm = boston['rm'].mean()
median_rm = boston['rm'].median()
std_rm = boston['rm'].std()

print('Mean (rm):', mean_rm)
print('Median (rm):', median_rm)
print('Standard Deviation (rm):', std_rm)
```

```python
# Exercice 5

titanic = pd.read_csv("titanic.csv")

titanic['Sex'] = titanic['Sex'].map({'female': 0, 'male': 1})

display(titanic[['Sex']].head())
```
