Exercice 3

Sleep Dataset

    Qualitative (valeurs textuelles ou categories) :

        Period -> texte, indique une periode (ex: "Annual")

        Type of Days -> categorie (ex: "All days")

        Age Group -> groupe d'age (ex: "15 years and over")

        Activity -> type d'activite (ex: "Sleeping")

        Sex -> categorie (ex: "Both")

    Quantitative (valeurs numeriques mesurables) :

        index -> identifiant numerique

        Year -> annee (nombre)

        Avg hrs per day sleeping -> nombre d'heures (valeur numerique)

        Standard Error -> mesure statistique (valeur numerique)

Mental Health Dataset

    Qualitative :

        Entity -> nom du pays (texte)

        Code -> code pays (texte court, ex: "AFG")

    Quantitative :

        index -> identifiant numerique

        Year -> annee (valeur numerique)

        Schizophrenia (%) -> pourcentage

        Bipolar disorder (%) -> pourcentage

        Eating disorders (%) -> pourcentage

        Anxiety disorders (%) -> pourcentage

        Drug use disorders (%) -> pourcentage

        Depression (%) -> pourcentage

        Alcohol use disorders (%) -> pourcentage
        (-> Tous sont des taux exprimes en nombres)

Credit Dataset

    Qualitative :

        Gender -> homme/femme (categorie)

        Married -> marie ou non (categorie)

        BankCustomer -> client ou non (categorie)

        Industry -> secteur d'emploi (texte)

        Ethnicity -> groupe ethnique (categorie)

        PriorDefault -> oui/non (categorie)

        Employed -> oui/non (categorie)

        DriversLicense -> oui/non (categorie)

        Citizen -> statut citoyen (categorie)

        ZipCode -> code postal (identifiant)

        Approved -> pret accorde ou non (categorie binaire)

    Quantitative :

        Age -> age en annees (nombre)

        Debt -> montant de dette (nombre)

        YearsEmployed -> annees d'emploi (nombre)

        CreditScore -> score de credit (nombre)

        Income -> revenu (nombre)
