```python
# Afficher la premi√®re et la derni√®re date
min_date = df['Order Date'].min()
max_date = df['Order Date'].max()

print(f"P√©riode couverte par le dataset : de {min_date.date()} √† {max_date.date()}")
```

```python
# V√©rifier les valeurs manquantes
print('\nValeurs manquantes par colonne :')
print(df.isnull().sum())

# V√©rifier le nombre de doublons
print('\nNombre de doublons :', df.duplicated().sum())

# Supprimer les doublons
df.drop_duplicates(inplace=True)
print('\nDoublons supprim√©s.')

# Aper√ßu apr√®s nettoyage
print('\nInfo apr√®s nettoyage :')
print(df.info())
```

```python
# S'assurer que la colonne est bien au format datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Trier par date
df_sorted = df.sort_values(by='Order Date')

# Calcul des √©carts entre deux commandes cons√©cutives
date_diffs = df_sorted['Order Date'].diff()

# Afficher les 10 plus gros √©carts de temps
print("\nPlus gros √©carts entre deux commandes :")
print(date_diffs.sort_values(ascending=False).head(10))
```

```python
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100

df['Order Year'] = df['Order Date'].dt.year

df['Order Month'] = df['Order Date'].dt.month

print(df[['Sales', 'Profit', 'Profit Margin', 'Order Date', 'Order Year', 'Order Month']].head())
```

```python
df['Shipping Delay'] = (df['Ship Date'] - df['Order Date']).dt.days

df['Sales Size'] = pd.cut(df['Sales'], bins=[0, 100, 500, 1000, df['Sales'].max()],
                              labels=['Small', 'Medium', 'Large', 'Very Large'])
df['Is Profitable'] = df['Profit'] > 0

print(df[['Sales','Shipping Delay', 'Sales Size','Profit Margin', 'Is Profitable']].head())
```

üîÑ √âtape 1 : df.groupby('Order Month')

    On regroupe toutes les lignes du DataFrame selon la valeur de la colonne 'Order Month' (qui contient des chiffres de 1 √† 12).

    Par exemple, toutes les ventes de janvier seront regroup√©es ensemble, puis f√©vrier, etc.

üì¶ √âtape 2 : ['Sales'].sum()

    Pour chaque mois, on somme les ventes (Sales) : c‚Äôest une agr√©gation.

    R√©sultat : un tableau avec 12 lignes, une par mois, contenant le total des ventes du mois.

üî¢ √âtape 3 : .sort_index()

    S'assure que les mois sont dans l'ordre num√©rique (1 ‚Üí 12), au cas o√π ils auraient √©t√© m√©lang√©s.

```python
monthly_sales = df.groupby('Order Month')['Sales'].sum().sort_index()

month_names = [calendar.month_name[i] for i in monthly_sales.index]

plt.figure(figsize=(15, 8))
plt.plot(month_names, monthly_sales.values, color='green', linestyle='dashed',
         linewidth=2, marker='o', markerfacecolor='blue', markersize=8)

plt.title('Total Sales by Month (All Years Combined)', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Month'] = df['Order Date'].dt.month
df['Category'] = df['Category'].astype(str)

categories = sorted(df['Category'].unique())

def plot_sales_by_category(category):
    df_filtered = df[df['Category'] == category]
    
    monthly_sales = df_filtered.groupby('Order Month')['Sales'].sum().sort_index()
    
    month_names = [calendar.month_name[i] for i in monthly_sales.index]
    
    plt.figure(figsize=(10, 5))
    plt.plot(month_names, monthly_sales.values, marker='o', color='teal', linestyle='-')
    plt.title(f"Total Sales by Month ‚Äì Category: {category}", fontsize=14)
    plt.xlabel("Month")
    plt.ylabel("Total Sales")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

interact(plot_sales_by_category, category=widgets.Dropdown(options=categories, description='Category:'))
```

```python
monthly_sales = df.groupby('State')['Sales'].sum().sort_index()

plt.figure(figsize=(15, 6))
plt.bar(monthly_sales.index, monthly_sales.values)
plt.title('Sales by State')
plt.xlabel('States')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

max_states = len(state_sales)

def plot_top_states(top_n):
    top_states = state_sales.head(top_n)
    
    plt.figure(figsize=(12, 6))
    plt.bar(top_states.index, top_states.values, color='skyblue')
    plt.title(f"Top {top_n} States by Sales", fontsize=14)
    plt.xlabel("State")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=75)
    plt.tight_layout()
    plt.show()

# Slider interactif
interact(plot_top_states, top_n=widgets.IntSlider(min=5, max=max_states, step=1, value=10, description='Top N:'))
```

```python
top_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

top_products_df = top_products.reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Profit',y='Product Name',hue='Product Name',data=top_products_df,palette='viridis',dodge=False,legend=False)

for i, value in enumerate(top_products_df['Profit']):
    plt.text(value, i, f"${value:,.2f}", va='center')

plt.title('Top 10 Most Profitable Products')
plt.xlabel('Total Profit')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category', alpha=0.6)
sns.regplot(data=df, x='Discount', y='Profit', scatter=False, color='black', line_kws={"linewidth": 2})

plt.title('Relationship Between Discount and Profit')
plt.xlabel('Discount Rate')
plt.ylabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
state_list = sorted(df['State'].unique())

def plot_state_dashboard(state):
    df_state = df[df['State'] == state]

    fig, axs = plt.subplots(1, 2, figsize=(16, 5))

    sales_by_category = df_state.groupby('Category')['Sales'].sum().sort_values()
    axs[0].barh(sales_by_category.index, sales_by_category.values, color='orange')
    axs[0].set_title(f"Sales by Category in {state}")
    axs[0].set_xlabel('Sales')

    profits_by_sub = df_state.groupby('Sub-Category')['Profit'].sum().sort_values()
    axs[1].barh(profits_by_sub.index, profits_by_sub.values, color='teal')
    axs[1].set_title(f"Profit by Sub-Category in {state}")
    axs[1].set_xlabel('Profit')

    plt.tight_layout()
    plt.show()

interact(plot_state_dashboard, state=widgets.Dropdown(options=state_list, description='State:'))
```
