```python
#Exercie 5

df = pd.read_csv("StudentHealth.csv")

# suppression des lignes o√π les colonnes concern√©es sont vides
df = df.dropna(subset=['Do you have Anxiety?', 'Choose your gender'])

# yes √† la question sur l'anxi√©t√©
anxious = df[df['Do you have Anxiety?'].str.lower() == 'yes']

# genre
gender_counts = anxious['Choose your gender'].value_counts()
""" print(gender_counts) """

# on convertit en proportion
gender_proportions = gender_counts / gender_counts.sum()
""" print(gender_proportions) """

plot_df = gender_proportions.reset_index()
plot_df.columns = ['Gender', 'Proportion']
plot_df['Proportion'] = plot_df['Proportion']*100

sns.barplot(data=plot_df, x='Gender', y='Proportion', hue='Gender', palette='Set2', legend=False)
plt.title("Pourcentage d'anxieux par genre")
plt.xlabel('Genre')
plt.ylabel('Pourcentage')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
# exercice 6
df = pd.read_csv("StudentHealth.csv")

print(df['Age'].describe())
print(df['Age'].unique())

# suppression lignes vides √©ventuelles
df = df.dropna(subset=['Age', 'Do you have Panic attack?'])

# conversion
df['Panic'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

# tranches d'√¢ge
bins = [17, 19, 21, 23, 25]
labels = ['18-19 ans', '20-21 ans', '22-23 ans', '24 ans +']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# moyenne des crises
grouped = df.groupby('Age Group')['Panic'].mean().reset_index()

grouped['Panic'] = grouped['Panic'] * 100
sns.barplot(data=grouped, x='Age Group', y='Panic', palette='viridis')
plt.title("Pourcentage d'√©tudiants ayant des crises de panique par tranche d'√¢ge")
plt.xlabel("Tranche d'√¢ge")
plt.ylabel("Pourcentage (%)")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
