```python
# tic tac toe

from typing import List

# afficher la grille
def display_board(board: List[str]) -> None:
    print("\n" + "\n-----------\n".join(
        [" | ".join(board[i:i+3]) for i in range(0, 9, 3)]
    ) + "\n")

# recuperer le coup joueur
def player_input(board: List[str], player: str) -> int:
    while True:
        try:
            position = int(input(f"Joueur {player}, choisissez une case (1-9) : ")) - 1
            if board[position] not in ('X', 'O'):
                return position
            print("Cette case est dÃ©jÃ  occupÃ©e. Essayez encore.")
        except (ValueError, IndexError):
            print("EntrÃ©e invalide. Veuillez choisir un numÃ©ro entre 1 et 9.")

# check gagnant
def check_win(board: List[str], player: str) -> bool:
    winning_combinations = [
        (0,1,2), (3,4,5), (6,7,8),
        (0,3,6), (1,4,7), (2,5,8),
        (0,4,8), (2,4,6)
    ]
    return any(all(board[i] == player for i in combo) for combo in winning_combinations)

def is_draw(board: List[str]) -> bool:
    return all(cell in ('X', 'O') for cell in board)

def play():
    board = [str(i+1) for i in range(9)]
    current_player = 'X'
    display_board(board)

    while True:
        pos = player_input(board, current_player)
        board[pos] = current_player
        display_board(board)

        if check_win(board, current_player):
            print(f"ğŸ‰ Le joueur {current_player} a gagnÃ© !")
            break
        elif is_draw(board):
            print("ğŸ¤ Match nul !")
            break

        current_player = 'O' if current_player == 'X' else 'X'

if __name__ == "__main__":
    print("Bienvenue dans le jeu du Morpion !\n")
    play()
```
