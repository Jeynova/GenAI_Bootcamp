Instructions

    The game is played on a grid that‚Äôs 3 squares by 3 squares.
    Players take turns putting their marks (O or X) in empty squares.
    The first player to get 3 of their marks in a row (up, down, across, or diagonally) is the winner.
    When all 9 squares are full, the game is over. If no player has 3 marks in a row, the game ends in a tie.


Hint

To do this project, you basically need to create four functions:

    display_board() ‚Äì To display the Tic Tac Toe board (GUI).
    player_input(player) ‚Äì To get the position from the player.
    check_win() ‚Äì To check whether there is a winner or not.
    play() ‚Äì The main function, which calls all the functions created above.

Note: The 4 functions above are just an example. You can implement many more helper functions or choose a completely different appoach if you want.

```python
# tic tac toe

from typing import List

# afficher la grille
def display_board(board: List[str]):
    print()

    for i in range(0, 9, 3):
        ligne = board[i] + " | " + board[i+1] + " | " + board[i+2]
        print(ligne)

        if i < 6:
            print("-----------")
    print()
    
# recuperer le coup joueur
def player_input(board: List[str], player: str):
    while True:
        try:
            choix = input(f"Joueur {player}, choisissez une case (1-9) ou 'q' pour quitter : ")
            if choix.lower() == 'q':
                return None
            position = int(choix) - 1
            if board[position] not in ('X', 'O'):
                return position
            print("Cette case est d√©j√† occup√©e. Essayez encore.")
        except (ValueError, IndexError):
            print("Veuillez choisir un num√©ro entre 1 et 9 ou 'q' pour quitter.")

# check gagnant
def check_win(board: List[str], player: str):
    for combo in [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),
        (0, 3, 6), (1, 4, 7), (2, 5, 8),
        (0, 4, 8), (2, 4, 6)
    ]:
        if all(board[i] == player for i in combo):
            return True
    return False

def is_draw(board: List[str]):
    return all(cell in ('X', 'O') for cell in board)

def play():
    board = [str(i+1) for i in range(9)]
    current_player = 'X'
    display_board(board)

    while True:
        pos = player_input(board, current_player)
        if pos is None:
            break
        
        board[pos] = current_player
        display_board(board)

        if check_win(board, current_player):
            print(f"Le joueur {current_player} a gagn√© !")
            break
        elif is_draw(board):
            print("ü§ù Match nul !")
            break

        current_player = 'O' if current_player == 'X' else 'X'

if __name__ == "__main__":
    print("Bienvenue dans le jeu du Morpion !\n")
    play()
```
