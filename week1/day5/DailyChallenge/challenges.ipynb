Challenge 2 : Longest Word
Instructions

    Write a function that finds the longest word in a sentence. If two or more words are found, return the first longest word.
    Characters such as apostrophe, comma, period count as part of the word (e.g. O‚ÄôConnor is 8 characters long).

Examples

longest_word("Margaret's toy is a pretty doll.") ‚ûû "Margaret's"

longest_word("A thing of beauty is a joy forever.") ‚ûû "forever."

longest_word("Forgetfulness is by all means powerless!") ‚ûû "Forgetfulness"


```python
# importer
def sort_words_alpha(input_string):
    # couper mots
    words = [word.strip() for word in input_string.split(',')]
    # trier
    sorted_words = sorted(words)
    # rejoindre
    return ','.join(sorted_words)

def longest_word(sentence):
    # separer mots
    words = sentence.split()
    # chercher plus long
    return max(words, key=len)

def menu():
    while True:
        # afficher menu
        print("\n--- Menu ---")
        print("1. Trier mots alphabetiquement")
        print("2. Trouver mot le plus long")
        print("q. Quitter")

        # demander choix
        choice = input("Choix : ").lower()

        if choice == '1':
            # trier mots
            chaine = input("Entrer des mots separes par des virgules : ")
            print("Resultat :", sort_words_alpha(chaine))

        elif choice == '2':
            # trouver plus long
            phrase = input("Entrer une phrase : ")
            print("Mot le plus long :", longest_word(phrase))

        elif choice in ('q', 'quit'):
            # quitter
            print("Bye.")
            break

        else:
            print("Choix invalide.")

if __name__ == "__main__":
    menu()
```
