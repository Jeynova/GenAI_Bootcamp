Exercise 1 : Call History
Instructions

    Create a class called Phone. This class takes a parameter called phone_number. When instantiating an object create an attribute called call_history which value is an empty list.

    Add a method called call that takes both self and other_phone (i.e another Phone object) as parameters. The method should print a string stating who called who, and add this string to the phone‚Äôs call_history.

    Add a method called show_call_history. This method should print the call_history.

    Add another attribute called messages to your __init__() method which value is an empty list.

    Create a method called send_message which is similar to the call method. Each message should be saved as a dictionary with the following keys:
        to : the number of another Phone object
        from : your phone number (also a Phone object)
        content

    Create the following methods: show_outgoing_messages(self), show_incoming_messages(self), show_messages_from(self)

    Test your code !!!

```python
# XP Ninja
class Phone:
    def __init__(self, phone_number):
        # Initialiser le num√©ro de t√©l√©phone et les historiques
        self.phone_number = phone_number
        self.call_history = []
        self.messages = []

    def call(self, other_phone):
        # Simuler un appel et l‚Äôenregistrer dans l‚Äôhistorique
        call_log = f"{self.phone_number} called {other_phone.phone_number}"
        print(call_log)
        self.call_history.append(call_log)

    def show_call_history(self):
        # Afficher l‚Äôhistorique des appels
        print("Call History:")
        for entry in self.call_history:
            print(f"- {entry}")

    def send_message(self, other_phone, content):
        # Envoyer un message et l‚Äôenregistrer comme dictionnaire
        message = {
            "to": other_phone.phone_number,
            "from": self.phone_number,
            "content": content
        }
        self.messages.append(message)
        other_phone.messages.append(message)

    def show_outgoing_messages(self):
        # Afficher les messages envoy√©s
        print("Outgoing Messages:")
        for msg in self.messages:
            if msg["from"] == self.phone_number:
                print(f"To {msg['to']}: {msg['content']}")

    def show_incoming_messages(self):
        # Afficher les messages re√ßus
        print("Incoming Messages:")
        for msg in self.messages:
            if msg["to"] == self.phone_number:
                print(f"From {msg['from']}: {msg['content']}")

    def show_messages_from(self, number):
        # Afficher les messages re√ßus d‚Äôun num√©ro sp√©cifique
        print(f"Messages from {number}:")
        for msg in self.messages:
            if msg["from"] == number and msg["to"] == self.phone_number:
                print(f"- {msg['content']}")
                
phone1 = Phone("0654321022")
phone2 = Phone("0765432133")
phone3 = Phone("0678901244")

phone1.call(phone2)
phone2.call(phone1)

phone1.show_call_history()
print()

phone1.send_message(phone2, "Yo!")
phone2.send_message(phone1, "Salut!")
phone3.send_message(phone1, "Salut ! Tu vas bien ?")
phone1.send_message(phone3, "Hey, cava ?")

print()
phone1.show_outgoing_messages()
print()
phone1.show_incoming_messages()
print()
phone1.show_messages_from("0612093487")
```
