Exercise 3 : Restaurant Menu Manager
Instructions

The purpose of this exercise is to create a restaurant menu. The code will allow a manager to add and delete dishes.

    Create a python file called menu_manager.py.

    Create a class called MenuManager.

    Create a method __init__ that instantiates an attribute called menu. The menu attributes value will be all the current dishes (list of dictionaries). Each dish will be stored in a dictionary where the keys are name, price, spice level and gluten index (which value is a boolean).

Here is a list of the dishes currently on the menu:

    Soup – 10 – B - False
    Hamburger – 15 - A - True
    Salad – 18 - A - False
    French Fries – 5 - C - False
    Beef bourguignon– 25 - B - True

    Meaning: For the spice level:
       • A = not spicy,
       • B = a little spicy,
       • C = very spicy


The dishes provided above should be the value of the menu attribute.

    Create a method called add_item(name, price, spice, gluten). This method will add new dishes to the menu.

    Create a method called update_item(name, price, spice, gluten). This method checks whether a dish is in the menu, if the dish exists then update it. If not notify the manager that the dish is not in the menu.

    Create a method called remove_item(name). This method should check if the dish is in the menu, if the dish exists then delete it and print the updated dictionary. If not notify the manager that the dish is not in the menu.


```python
import math
import random


# Exercise 1

class Circle:
    def __init__(self, radius=1.0):
        # Initialiser le rayon du cercle
        self.radius = radius

    def get_area(self):
        # Calculer et retourner l'aire du cercle
        return math.pi * self.radius ** 2

    def get_perimeter(self):
        # Calculer et retourner le périmètre du cercle
        return 2 * math.pi * self.radius

    def describe(self):
        # Afficher la définition géométrique d’un cercle
        print("A circle is a shape with all points equidistant from the center.")

# Exercise 2

class MyList:
    def __init__(self, letters):
        # Initialiser la liste avec des lettres
        self.letters = letters

    def get_reversed(self):
        # Retourner la liste inversée
        return self.letters[::-1]

    def get_sorted(self):
        # Retourner la liste triée
        return sorted(self.letters)

    def get_random_numbers_list(self):
        # Générer une liste de nombres aléatoires de même longueur
        return [random.randint(0, 100) for _ in self.letters]

# Exercise 3: Restaurant Menu Manager

class MenuManager:
    def __init__(self):
        # Initialiser le menu avec des plats prédéfinis
        self.menu = [
            {"name": "Soup", "price": 10, "spice": "B", "gluten": False},
            {"name": "Hamburger", "price": 15, "spice": "A", "gluten": True},
            {"name": "Salad", "price": 18, "spice": "A", "gluten": False},
            {"name": "French Fries", "price": 5, "spice": "C", "gluten": False},
            {"name": "Beef bourguignon", "price": 25, "spice": "B", "gluten": True},
        ]

    def add_item(self, name, price, spice, gluten):
        # Ajouter un nouveau plat au menu
        self.menu.append({
            "name": name,
            "price": price,
            "spice": spice,
            "gluten": gluten
        })

    def update_item(self, name, price, spice, gluten):
        # Mettre à jour un plat existant, ou avertir s’il n’existe pas
        for dish in self.menu:
            if dish["name"].lower() == name.lower():
                dish.update({
                    "price": price,
                    "spice": spice,
                    "gluten": gluten
                })
                return
        print(f"[INFO] '{name}' is not in the menu.")

    def remove_item(self, name):
        # Supprimer un plat existant ou avertir s’il n’existe pas
        for dish in self.menu:
            if dish["name"].lower() == name.lower():
                self.menu.remove(dish)
                print("[INFO] Updated menu:")
                for d in self.menu:
                    print(d)
                return
        print(f"[INFO] '{name}' is not in the menu.")


c = Circle(2.5)
print("Circle area:", round(c.get_area(), 2))
print("Circle perimeter:", round(c.get_perimeter(), 2))
c.describe()
print()

ml = MyList(['b', 'a', 'd', 'c'])
print("Reversed list:", ml.get_reversed())
print("Sorted list:", ml.get_sorted())
print("Random numbers:", ml.get_random_numbers_list())
print()

manager = MenuManager()
manager.add_item("Pizza", 20, "C", True)
manager.update_item("Pizza", 22, "B", False)
manager.remove_item("Pizza")
manager.remove_item("Taco")
```
