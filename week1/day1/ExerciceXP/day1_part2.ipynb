```python

import random

def saison_par_mois():
    while True:
        mois = input("Entrez un numÃ©ro de mois (1 Ã  12) : ")
        if mois.isdigit():
            mois = int(mois)
            if 1 <= mois <= 12:
                if 3 <= mois <= 5:
                    print("C'est le printemps")
                elif 6 <= mois <= 8:
                    print("C'est l'Ã©tÃ©")
                elif 9 <= mois <= 11:
                    print("C'est l'automne")
                else:
                    print("C'est l'hiver")
                break
            else:
                print("Le nombre doit Ãªtre entre 1 et 12.")
        else:
            print("Veuillez entrer un nombre valide.")

def afficher_nombres():
    print("Nombres de 1 Ã  20 :")
    for i in range(1, 21):
        print(i)

    print("Nombres avec index pair :")
    for i in range(0, 20, 2):
        print(i + 1)

def verifier_nom():
    my_name = "Alex"
    while True:
        name = input("Entrez votre prÃ©nom : ").strip()
        if name.lower() == my_name.lower():
            print("PrÃ©nom reconnu.")
            break

def chercher_nom_dans_liste():
    names = ['Samus', 'Cortana', 'V', 'Link', 'Mario', 'Cortana', 'Samus']
    user_name = input("Entrez votre prÃ©nom : ").strip()

    for index, nom in enumerate(names):
        if nom.lower() == user_name.lower():
            print("Index trouvÃ© :", index)
            return
    print("PrÃ©nom non trouvÃ© dans la liste.")

def plus_grand_nombre():
    nombres = []
    for i in range(1, 4):
        while True:
            val = input(f"Entrez le nombre {i} : ")
            if val.strip().isdigit():
                nombres.append(int(val))
                break
            else:
                print("Veuillez entrer un entier valide.")
    print("Le plus grand nombre est :", max(nombres))

def jeu_devinette():
    wins = 0
    losses = 0

    while True:
        guess = input("Devinez un nombre entre 1 et 9 (ou tapez 'exit' pour quitter) : ")
        if guess.lower() == "exit":
            break
        if not guess.isdigit() or not 1 <= int(guess) <= 9:
            print("Veuillez entrer un nombre entre 1 et 9.")
            continue

        guess = int(guess)
        number = random.randint(1, 9)
        if guess == number:
            print("GagnÃ©.")
            wins += 1
        else:
            print(f"Perdu. Le bon nombre Ã©tait {number}.")
            losses += 1

    print(f"Parties gagnÃ©es : {wins}, Parties perdues : {losses}")



saison_par_mois()
afficher_nombres()
verifier_nom()
chercher_nom_dans_liste()
plus_grand_nombre()
jeu_devinette()
```
