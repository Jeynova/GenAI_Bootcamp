üåü Exercise 8 : Sandwich Orders

Using the list below :

sandwich_orders = ["Tuna sandwich", "Pastrami sandwich", "Avocado sandwich", "Pastrami sandwich", "Egg sandwich", "Chicken sandwich", "Pastrami sandwich"].

    The deli has run out of pastrami, use a while loop to remove all occurrences of Pastrami sandwich from sandwich_orders.

    We need to prepare the orders of the clients:

    Create an empty list called finished_sandwiches.

```python
def afficher_hello_world():
    print(("Hello world\n" * 4).strip())

def calcul_puissance():
    print("R√©sultat de (99 ** 3) * 8 :", (99 ** 3) * 8)

def comparer_noms():
    my_name = "J√©r√©my"
    name = input("Comment vous appelez-vous ? ").strip().capitalize()
    if name == my_name:
        print("Trop fort, on a le m√™me nom !")
    else:
        print(f"Enchant√© {name}. Je suis {my_name}.")

def verifier_taille():
    while True:
        taille = input("Entrez votre taille en cm : ")
        if taille.isdigit():
            taille = int(taille)
            if taille > 145:
                print("Vous pouvez monter !")
            else:
                print("D√©sol√©, vous √™tes trop petit.")
            break
        else:
            print("Veuillez entrer un nombre valide.")

def manipuler_sets():
    my_fav_numbers = {7, 42, 3}
    my_fav_numbers.update({9, 13})
    my_fav_numbers.discard(13)
    friend_fav_numbers = {1, 4, 7}
    our_fav_numbers = my_fav_numbers | friend_fav_numbers
    print("Nos nombres favoris :", our_fav_numbers)

def explication_tuple():
    my_tuple = (1, 2, 3)
    print(my_tuple)
    print("Les tuples sont immuables -> on ne peut pas modifier leur contenu apr√®s cr√©ation.")

def manipuler_liste_panier():
    basket = ["Banana", "Apples", "Oranges", "Blueberries"]
    basket.remove("Banana")
    basket.remove("Blueberries")
    basket.append("Kiwi")
    basket.insert(0, "Apples")
    print("Nombre de pommes :", basket.count("Apples"))
    basket.clear()
    print("Contenu final du panier :", basket)

def gestion_sandwiches():
    sandwich_orders = [
        "Tuna sandwich", "Pastrami sandwich", "Avocado sandwich", 
        "Pastrami sandwich", "Egg sandwich", "Chicken sandwich", 
        "Pastrami sandwich"
    ]

    print("D√©sol√©, plus de sandwich au pastrami.")
    sandwich_orders = [s for s in sandwich_orders if s != "Pastrami sandwich"]

    finished_sandwiches = []
    while sandwich_orders:
        sandwich = sandwich_orders.pop(0)
        print(f"Voici votre {sandwich.lower()}")
        finished_sandwiches.append(sandwich)

afficher_hello_world()
calcul_puissance()
comparer_noms()
verifier_taille()
manipuler_sets()
explication_tuple()
manipuler_liste_panier()
gestion_sandwiches()
```
