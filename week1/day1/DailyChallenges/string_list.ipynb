Challenge 2

Write a program that asks a string to the user, and display a new string with any duplicate consecutive letters removed.

Examples


user's word : "ppoeemm" â "poem"

user's word : "wiiiinnnnd" â "wind"

user's word : "ttiiitllleeee" â "title"

Notes

Final strings wonâ€™t include words with double letters (e.g. â€œpassingâ€, â€œlotteryâ€).

```python
def generate_multiples():
    while True:
        num_input = input("Entrez un nombre (ou 'q' pour quitter): ")
        if num_input.lower() == 'q':
            return
        try:
            number = int(num_input)
            break
        except ValueError:
            print("Veuillez entrer un nombre entier valide.")

    while True:
        length_input = input("Entrez une longueur (ou 'q' pour quitter): ")
        if length_input.lower() == 'q':
            return
        try:
            length = int(length_input)
            if length <= 0:
                print("Veuillez entrer un nombre positif.")
                continue
            break
        except ValueError:
            print("Veuillez entrer un nombre entier valide.")

    multiples = [number * i for i in range(1, length + 1)]
    print("Multiples :", multiples)


def remove_consecutive_duplicates():
    while True:
        text = input("Entrez un mot (ou 'q' pour quitter): ")
        if text.lower() == 'q':
            return
        if text.isalpha():
            break
        print("Veuillez entrer uniquement des lettres.")

    result = text[0] if text else ""
    for char in text[1:]:
        if char != result[-1]:
            result += char
    print("Mot :", result)


def prog():
    while True:
        print("\nMenu:")
        print("1. GÃ©nÃ©rer des multiples")
        print("2. Retirer les caractÃ¨res qui se suivent")
        print("3. Quitter")
        choice = input("Choisissez 1,2 ou 3: ")

        match choice:
            case '1':
                generate_multiples()
            case '2':
                remove_consecutive_duplicates()
            case '3':
                print("Au revoir !")
                break
            case _:
                print("Veuillez entrer 1, 2 ou 3.")

prog()
```
