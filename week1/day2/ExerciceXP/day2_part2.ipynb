```python
# Exercice 1 : Recherche d'anniversaire avec contrôle d'input

def birthday_lookup():
    birthdays = {
        "Alice": "1990/04/18",
        "Bob": "1988/09/12",
        "Charlie": "1995/02/26",
        "Diana": "1992/06/07",
        "Eva": "2000/11/30"
    }
    print("Bienvenue ! Tu peux chercher la date d'anniversaire d'une personne dans la liste.")
    while True:
        person_name = input("Entre un prénom (ou 'quitter' pour sortir) : ").strip().capitalize()
        if person_name.lower() == "quitter":
            print("Fin de la recherche.")
            break
        if person_name in birthdays:
            print(f"L'anniversaire de {person_name} est le {birthdays[person_name]}.")
        else:
            print(f"Désolé, pas d'information pour {person_name}.")

birthday_lookup()


# Exercice 2 : Recherche avancée d'anniversaire avec set et input sécurisé

def birthday_lookup_advanced():
    birthdays = {
        "Alice": "1990/04/18",
        "Bob": "1988/09/12",
        "Charlie": "1995/02/26",
        "Diana": "1992/06/07",
        "Eva": "2000/11/30"
    }
    # Utilisation de set pour affichage unique et rapide si la liste devient longue
    all_names = set(birthdays.keys())
    print("Voici la liste des personnes disponibles :")
    print(", ".join(sorted(all_names)))
    while True:
        person_name = input("Qui veux-tu chercher ? (ou 'quitter') : ").strip().capitalize()
        if person_name.lower() == "quitter":
            print("Fin de la recherche.")
            break
        if person_name in all_names:
            print(f"L'anniversaire de {person_name} est le {birthdays[person_name]}.")
        else:
            print(f"Désolé, nous n'avons pas l'anniversaire de {person_name}.")

birthday_lookup_advanced()


# Exercice 3 : Vérifier l'index, gestion robustesse, index multiples avec list + enumerate + filter

def check_name_index():
    names = ['Samus', 'Cortana', 'V', 'Link', 'Mario', 'Cortana', 'Samus']
    user_name = input("Quel est ton prénom ? ").strip().capitalize()
    # Filtre tous les index du prénom
    indices = list(map(lambda t: t[0], filter(lambda t: t[1] == user_name, enumerate(names))))
    if indices:
        print(f"{user_name} apparaît à l'index {indices[0]} (première occurrence).")
        if len(indices) > 1:
            print(f"Toutes les occurrences : {indices}")
    else:
        print("Ton prénom n'est pas dans la liste.")

check_name_index()


# Exercice 4 : Double Dés avec dictionnaire pour stats + lambda, map

import random

def throw_dice():
    # Lancer un dé (1 à 6)
    return random.randint(1, 6)

def throw_until_doubles():
    # Lance les deux dés jusqu'à obtenir deux fois la même valeur
    throws = 0
    while True:
        die1, die2 = throw_dice(), throw_dice()
        throws += 1
        if die1 == die2:
            break
    return throws

def main_double_dice():
    num_trials = 100
    throws_needed_list = [throw_until_doubles() for _ in range(num_trials)]
    total_throws = sum(throws_needed_list)
    average_throws = round(total_throws / num_trials, 2)
    # Statistique avancée : combien de parties ont pris plus de 5 lancers ?
    many_throws_count = len(list(filter(lambda x: x > 5, throws_needed_list)))
    # Dictionnaire pour compter combien de parties ont pris x lancers
    from collections import Counter
    throws_stats = dict(Counter(throws_needed_list))
    print(f"Nombre total de lancers pour obtenir {num_trials} doubles : {total_throws}")
    print(f"Nombre moyen de lancers pour obtenir un double : {average_throws}")
    print(f"Nombre de parties avec plus de 5 lancers : {many_throws_count}")
    print("Détail du nombre de lancers pour chaque double :")
    print(throws_stats)

main_double_dice()
```
