This challenge is about Biology that will put emphasis on your knowledge of classes, inheritance and polymorphism.

    Build a DNA object. DNA is composed of chromosomes which is itself composed of Genes.
        A Gene is a single value 0 or 1, it can mutate (flip).
        A Chromosome is a series of 10 Genes. It also can mutate, meaning a random number of genes can randomly flip (1/2 chance to flip).
        A DNA is a series of 10 chromosomes, and it can also mutate the same way Chromosomes can mutate.

    Implement these classes as you see fit.

    Create a new class called Organism that accepts a DNA object and an environment parameter that sets the probability for its DNA to mutate.

    Instantiate a number of Organism and let them mutate until one gets to a DNA which is only made of 1s. Then stop and record the number of generations (iterations) it took.

Write your results in you personal biology research notebook and tell us your conclusion :).

```python
import random
from typing import List

# Classe gene
class Gene:
    def __init__(self, value: int = None):
        # Initialise la valeur
        self.value = value if value is not None else random.randint(0, 1)

    # muter
    def mutate(self):
        self.value = 1 - self.value

    # print valeur
    def __repr__(self):
        return str(self.value)

# Chromosme
class Chromosome:
    def __init__(self, genes: List[Gene] = None):
        # Initialiser les genes
        self.genes = genes if genes else [Gene() for _ in range(10)]

    # Mutater les chromosomes
    def mutate(self):
        for gene in self.genes:
            if random.random() < 0.5:
                gene.mutate()

    # Vérification
    def is_perfect(self):
        return all(g.value == 1 for g in self.genes)

    # Représentation textuelle
    def __repr__(self):
        return ''.join(map(str, self.genes))

# Adn
class DNA:
    def __init__(self, chromosomes: List[Chromosome] = None):
        self.chromosomes = chromosomes if chromosomes else [Chromosome() for _ in range(10)]

    # Mutater l adn
    def mutate(self):
        for chromosome in self.chromosomes:
            if random.random() < 0.5:
                chromosome.mutate()

    # Vérifier l adn
    def is_perfect(self):
        return all(chrom.is_perfect() for chrom in self.chromosomes)

    def __repr__(self):
        return '\n'.join(map(str, self.chromosomes))

# organisme
class Organism:
    def __init__(self, dna: DNA, environment_factor: float):
        self.dna = dna
        self.environment_factor = environment_factor

    # muter avec environnement
    def mutate(self):
        if random.random() < self.environment_factor:
            self.dna.mutate()

    # Vérifier l organisme
    def is_perfect(self):
        return self.dna.is_perfect()

# Run simu
def run_simulation(environment_factor: float = 0.8, population_size: int = 1) -> int:
    # Initialise la population
    population = [Organism(DNA(), environment_factor) for _ in range(population_size)]
    generations = 0

    while True:
        generations += 1
        for organism in population:
            organism.mutate()
            if organism.is_perfect():
                return generations

if __name__ == "__main__":
    print("Simulation de mutation génétique")
    generations_needed = run_simulation(environment_factor=0.9)
    print(f"Nombre de générations nécessaires : {generations_needed}")
```
