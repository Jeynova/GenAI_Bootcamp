```python
def add_two_numbers(a, b):
    if not isinstance(a, int) or not isinstance(b, int):
        return "Veuillez entrer des nombres"
    
    return a + b

print(add_two_numbers(5, 3))
print(add_two_numbers(10, "5"))

def greet(name):
    return f"Hello, {name}"

print(greet(input("Entrez un nom")))

def check_even_odd(num):
    if not isinstance(num, int):
        return "Veuillez entrer un nombre"
    
    if num % 2 == 0:
        return "Even"
    
    return "Odd"

print(check_even_odd(2))
print(check_even_odd(3))

def sum_list(numbers):
    if not isinstance(numbers, list):
        return "Veuillez entrer un tableau"
    
    for num in numbers:
        if not isinstance(num, (int, float)):
            return "Tout les elements doivent être des nombres"
    
    return sum(numbers)

print(sum_list([1, 2, 3, 4]))
print(sum_list([1, -1, 0.5]))
print(sum_list("1234"))
print(sum_list([1, "two", 3]))

def print_days():
    days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
    for day in days:
        print(day)

print_days()

def check_sign(num):
    if num > 0:
        return "Positive"
    elif num == 0:
        return "Zero"
    
    return "Negative"

print(check_sign(-12))
print(check_sign(0))
print(check_sign(6))

def repeat_word(word,repeat):
    for _ in range(repeat):
        print(word)

repeat_word("Coucou",3)

def find_largest(num):
    if not isinstance(num, list):
        return "Veuillez entrer une liste"
    
    for item in num:
        if not isinstance(item, (int, float)):
            return "Votre liste ne doit contenir que des nombres"
    
    return max(num)

# Example usage
print(find_largest([3, 6, 1, 367, 123, 8]))
print(find_largest([3, "hello", 5]))

def check_letter(word,letter):
    if letter in word:
        return "True"
    return "False"

print(check_letter("Banana","e"))
print(check_letter("Apple","p"))
```

```python
def count_to_number(n):
    if not isinstance(n, (int, float)):
        return print("Veuillez entrer un nombre.")

    n = int(n)
    rangeNum = [0] if n == 0 else range(1, n + 1) if n > 0 else range(1, n - 1, -1)
    
    for num in rangeNum:
        print(num)


print("Essai 1")
count_to_number("test")
print("Essai 2")
count_to_number(-3)
print("Essai 3")
count_to_number(0)
print("Essai 4")
count_to_number(5)
```
