```python
def calculatePetYears(humanYears):

    if not isinstance(humanYears, (int, float)):
        return "Veuillez entrer un nombre"

    humanYears = int(humanYears)

    if humanYears == 1:
        catYears = 15
        dogYears = 15
    elif humanYears == 2:
        catYears = 15 + 9
        dogYears = 15 + 9
    else:
        catYears = 15 + 9 + (humanYears - 2) * 4
        dogYears = 15 + 9 + (humanYears - 2) * 5

    return [humanYears, catYears, dogYears]

print(calculatePetYears(1))
print(calculatePetYears(2))
print(calculatePetYears(3))
print(calculatePetYears(10))

```
