Think Like a Data Scientist - Your First ML Blueprint
The Mission
You are a junior data analyst at a popular online education platform. The company wants to be proactive and help students who might be struggling.
Your mission is to design a plan (a blueprint) for a machine learning model that can predict if a student is at risk of failing their final exam. You don't need to write any code. You just need to make the decisions a data scientist would make at the start of a project.
Part 1: Framing the Problem (What are we even doing?)
Before a data scientist writes a single line of code, they must understand the problem.
Instructions:
Based on your mission (predicting "Pass" or "Fail"), which of the three main types of machine learning is the most appropriate for this task?
Supervised Learning (predicting a known outcome)
Unsupervised Learning (finding hidden groups)
Reinforcement Learning (learning through trial and error)
Explain your choice in one sentence. (Hint: Do you have a specific target you are trying to predict?)
Part 2: The Detective Work (Feature Selection)
A model is only as good as the data it learns from. Your next job is to think about what data you would need. These pieces of data are called "features."
Instructions:
Brainstorm: List at least 5 potential features you think would be useful for predicting a student's exam result. Think about all the data an online platform might collect.
Guiding questions: What behaviors indicate a student is engaged? What shows they are struggling? Think beyond just "time spent on the site."
Prioritize: From your list, choose the top 3 most powerful features. For each one, write a single sentence explaining why you believe it's a strong predictor.
Example Format:
Feature: Number of practice quizzes completed.
Why it's powerful: This directly measures how much a student is practicing and testing their knowledge.
Part 3: Choosing Your Tool (Model Selection)
Now that you have your data ideas, you need to think about the type of algorithm you would use.
Instructions:
In Supervised Learning, there are two main families of models:
Classification Models: Predict a category or a label (e.g., Spam or Not Spam, Cat or Dog, Pass or Fail).
Regression Models: Predict a continuous numerical value (e.g., the price of a house, the temperature tomorrow).
Which family of models is the right one for your mission? Explain why in one sentence.
Part 4: The Reality Check (Model Evaluation)
Your model is built! Now, how do you know if it's any good? More importantly, how do you explain its errors?
Instructions:
Imagine your model makes two types of mistakes:
Mistake A (False Negative): The model predicts a student will PASS, but they actually FAIL. The platform does nothing, and the student fails without receiving any help.
Mistake B (False Positive): The model predicts a student will FAIL, but they would have actually PASSED. The platform intervenes, offering them extra tutoring and resources.
Answer this critical question:
Which of these two mistakes is worse for the student and the company? Justify your answer. This decision will determine how you ultimately evaluate and tune your model.

Projet Machine Learning : Prédiction de réussite étudiante
1. Problème et type d'apprentissage

Objectif : Prédire si un étudiant réussira (pass) ou échouera (fail) à son examen final.
Type d'apprentissage choisi : Apprentissage supervisé

On essaie de prédire un résultat bien défini à partir de données existantes. Il s’agit donc d’un problème de classification supervisée.

2. Sélection des variables importantes (features)

Variable 1 : Nombre de modules complétés sur la plateforme
Plus un étudiant suit de modules, plus il est engagé, ce qui augmente ses chances de réussir.

Variable 2 : Score moyen aux quizzs réalisés en ligne
Cela donne une bonne idée de sa compréhension générale des cours.

Variable 3 : Nombre de jours consécutifs sans activité
De longues périodes sans se connecter peuvent signaler un désengagement ou des difficultés, et donc un risque plus élevé d’échec.

3. Type de modèle choisi

Famille de modèle choisie : Classification
Le modèle doit prédire une étiquette (pass ou fail), ce qui correspond à une tâche de classification et non de régression.

4. Stratégie d’évaluation

Erreur la plus critique : Erreur de type A (faux négatif)

C’est plus grave de penser qu’un étudiant va réussir alors qu’il échoue, car il ne recevra aucune aide. Même si on aide par erreur un étudiant qui aurait réussi (faux positif), cela n’a pas de conséquence négative sérieuse, au contraire. Il vaut donc mieux prévenir que guérir.
